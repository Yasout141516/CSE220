```python
# Complete the functions defined in this cell


# Test 01: Play Right
def playRight(sequence,beats):
  
  for i in beats:
    if i==1:
      shiftRight(sequence)
  return sequence
def shiftRight(a):
    temp=a[len(a)-1]
    for i in range(len(a)-1,-1,-1):
      a[i]=a[i-1]
    a[0]=temp
    return a


# Test 02: Discard Cards
def discardCards(cards,number):
  for i in range(len(cards)-1,-1,-1):
    if cards[i]==number:
      
      shiftLeft(cards,i)
  return cards
def shiftLeft(a,idx):
  for i in range(idx+1,len(a)):
    a[i-1]=a[i]
  a[len(a)-1]=0
  return a
  


#Test 03: Merge Lineup
def mergeLineup(pokemon_1, pokemon_2):
  team=np.zeros(len(pokemon_1),dtype=np.int64)
  left=0
  right=len(pokemon_1)-1
  for i in range(len(pokemon_1)):
    if pokemon_2[right]==None:
      pokemon_2[right]=0
    if pokemon_1[left]==None:
      pokemon_1[left]=0
    team[i]=pokemon_1[left]+pokemon_2[right]
    right-=1
    left+=1
  return team


# Test 04: Balance your Salami
def balanceSalami(a):
  leftsum=0
  rightsum=0
  r=len(a)-1
  l=0
  for i in range(len(a)):
    if l>r:
      return compare(leftsum,rightsum)
    else:
      leftsum+=a[l]
      rightsum+=a[r]
      l+=1
      r=-1
def compare(leftsum,rightsum):
  if leftsum==rightsum:
    return True
  else:
    return False
  

# Test 05: Protecc Salami
print("///  Test 05: Protecc Salami  ///")
def protectSalami(a):
  max=high(a)
  new=np.zeros(max+1,dtype=np.int64)
  for i in a:
    for j in range(len(new)):
      if i==j:
        new[i]+=1
  
  return comparing(new)
def comparing(new):
  count=0
  for i in range(len(new)):
    if new[i]>1:
      for j in range(i+1,len(new)):      
        if new[i]==new[j]:
          count+=1
  if count>=1:   
    return True
  else:
    return False
      
def high(a):
  max=a[0]
  for i in a:
    if max<i:
      max=i
  return max


# Test 06: Odd Even Wave
def waveYourFlag(a):
  new=np.zeros(len(a),dtype=np.int64)
  if a[0]%2==0:
    odd=1
    even=0
  else:
    odd=0
    even=1
  for i in range(len(a)):
    if a[i]%2==0:
      new[even]=a[i]
      even+=2
    else:
      new[odd]=a[i]
      odd+=2
  return new





```
